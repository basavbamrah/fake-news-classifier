```python
X= df.drop('label', axis=1)
```

```python
Y = df['label']
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
```

```python
vocab_size = 5000
```

```python
msg = X.copy()
msg.reset_index(drop=True, inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
corpus = []

for i in range(len(msg)):
    review = re.sub('[^a-zA-Z]', ' ', msg['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    
    corpus.append(review)
```

```python
corpus
```

```python
onehot_rep = [one_hot(word, vocab_size) for word in corpus]
```

```python
onehot_rep
```

```python
sent_len = 20

embedding_doc = pad_sequences(onehot_rep, maxlen=sent_len, padding='pre')
```

```python
embedding_doc
```

```python
embeding_vector_length = 40

model = Sequential()
model.add(Embedding(vocab_size, embeding_vector_length, input_length=sent_len))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
import numpy as np

X= np.array(embedding_doc)
y= np.array(Y)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=.33)
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_test, y_test))
```

```python
model.save('model_title.h5')
```

```python
y_pred = model.predict(X_test)
```

```python
y_test
```

```python
y_pred = [0 if x < 0.5 else 1 for x in y_pred]
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
